# 2 Análise descritiva

```python
# Valor_previsto total por hotel
df_reserva = pd.read_csv('f_reservanovo.csv', sep=';')
df_reserva
```

```python
# Unificando tabelas df_hotel e df_reserva
df_total_hotel = pd.merge(df_reserva, df_hotel, on='hotel_id', how='inner')
df_total_hotel
```

```python
# Verificando tipo de moedas
df_total_hotel['moeda'].unique()
```

Como há pagamentos em outros tipos de moeda, será realizada a conversão para Reais(BRL) utilizando a seguinte taxa:

Dólar(USD) = 5.30 BRL

Euro(EUR) = 6.30 BRL

(Cotação do dia 20/01/2026)

```python
usd = 5.30
eur = 6.30
df_total_hotel['valor_previsto_corrigido'] = df_total_hotel['valor_previsto']
df_total_hotel.loc[df_total_hotel['moeda'] == 'USD', 'valor_previsto_corrigido'] = df_total_hotel['valor_previsto'] * usd
df_total_hotel.loc[df_total_hotel['moeda'] == 'EUR', 'valor_previsto_corrigido'] = df_total_hotel['valor_previsto'] * eur
```

```python
# Criando tabela somente de diárias realizadas
df_total_hotel_finalizada = df_total_hotel.loc[df_total_hotel['status'] == 'Finalizada']
df_total_hotel_finalizada.head()
```

```python
plt.figure(figsize=(10,6))
receita_hoteis = df_total_hotel_finalizada.groupby('nome')['valor_previsto_corrigido'].sum().reset_index().sort_values(by= 'valor_previsto_corrigido', ascending=False)
sns.barplot(data=receita_hoteis, x= 'nome', y= 'valor_previsto_corrigido')
plt.title('Valor por hotel (em Milhões)')
plt.ylabel('Valor previsto (em Reais)')
plt.xlabel('Hotel')
plt.show()
```

#### Quais são as métricas a partir do ADR e receita diária?

```python
# Calculando estatística descritiva da receita das diárias a partir do tipo de quarto
df_diaria_novo['tipo_quarto'].unique()
```

```python
quartos = ['Standard', 'Superior', 'Deluxe', 'Suite', 'Familia']
```

```python
for quarto in quartos:
    array_receita = np.array(df_diaria_novo['receita_diarias'].loc[df_diaria_novo['tipo_quarto'] == quarto])
    maximo = np.max(array_receita)
    minimo = np.min(array_receita)
    media = np.mean(array_receita)
    mediana = np.median(array_receita)
    q1 = np.quantile(array_receita, 0.25)
    q3 = np.quantile(array_receita, 0.75)
    iqr = q3 - q1
    ls = q3 + iqr*1.5
    li = q1 - iqr*1.5
    outliersuperior = df_diaria_novo['receita_diarias'].loc[(df_diaria_novo['tipo_quarto'] == quarto) & (df_diaria_novo['receita_diarias'] > ls)]
    outlierinferior = df_diaria_novo['receita_diarias'].loc[(df_diaria_novo['tipo_quarto'] == quarto) & (df_diaria_novo['receita_diarias'] < li)]
    print(f'------------Informações: Quarto tipo {quarto}------------')
    print(f'A maior receita diária foi: {maximo:.2f} Reais.')
    print(f'A menor receita diária foi: {minimo:.2f} Reais.')
    print(f'A receita diária média foi: {media:.2f} Reais.')
    print(f'A mediana da receita diária é: {mediana:.2f} Reais.')
    print(f'25% da receita diária está em: {q1:.2f} Reais.')
    print(f'75% da receita diária está em: {q3:.2f} Reais.')
    print(f'Há {len(outliersuperior)} quartos do tipo {quarto} a cima da média de receita diária (outlier superior)')
    print(f'Há {len(outlierinferior)} quartos do tipo {quarto} abaixo da média de receita diária (outlier inferior)')
    print('')
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df_diaria_novo, x='tipo_quarto', y='receita_diarias', order = quartos)
plt.title('Distribuição de Receita das diárias por Tipo de Quarto reservado')
plt.xlabel('Receita das diárias')
plt.ylabel('Tipo de Quarto')
plt.ylim(400, 25000)
plt.show()
```

#### Existe correlação entre a nota final e a nota da limpeza ou atendimento?

```python
df_reserva_novo = pd.read_csv('f_reservanovo.csv', sep=';')
df_nota_reserva = pd.merge(df_review_ratio, df_reserva_novo, on='reserva_id', how='inner')
df_nota_reserva
```

```python
# Calculando correlação entre nota final e nota de limpeza
plt.figure(figsize=(20,10))
sns.lmplot(data=df_nota_reserva, x='limpeza', y='nota')
plt.show()
x = np.array(df_nota_reserva['limpeza'])
y = np.array(df_nota_reserva['nota'])
correlacao = np.corrcoef(x, y)
coef = correlacao[0,1]
print(coef)
```

Com o índice de correlação corrigido igual a 0,82, podemos afirmar que há uma forte correlação entre a nota da limpeza e a nota final.

```python
# Calculando correlação entre nota final e nota de atendimento
plt.figure(figsize=(20,10))
sns.lmplot(data=df_nota_reserva, x='atendimento', y='nota')
plt.show()
x = np.array(df_nota_reserva['atendimento'])
correlacao = np.corrcoef(x, y)
coef = correlacao[0,1]
print(coef)
```

Com o índice de correlação corrigido igual a 0,78, podemos afirmar que há uma forte correlação entre a nota de atendimento e a nota final.

```python
# Calculando correlação entre nota final e nota de localização
plt.figure(figsize=(20,10))
sns.lmplot(data=df_nota_reserva, x='localizacao', y='nota')
plt.show()
x = np.array(df_nota_reserva['localizacao'])
correlacao = np.corrcoef(x, y)
coef = correlacao[0,1]
print(coef)
```

Com o índice de correlação corrigido igual a 0,85, podemos afirmar que há uma forte correlação entre a nota de localização e a nota final.
